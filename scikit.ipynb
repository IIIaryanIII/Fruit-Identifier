```python
from sklearn import tree
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
features = [[150, 0],
            [170, 0],
           [130, 1],
           [120, 1]]
```

```python
labels = ["apples", "apples", "orange", "orange"]
```

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(features, labels)
```

```python
prediction = clf.predict([[108, 2]])
```

```python
print(f"The fruit is {prediction[0]}")
```

```python
from sklearn.tree import plot_tree
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(11,200))
```

```python
plot_tree(clf, feature_names=["weight", "colour"], class_names=["apple", "orange"], filled=True)
```
